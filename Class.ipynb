```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import expon
```

```python
class Plot():
    
    def __init__(self, fname1, fname2, fname3, data):
        
        self.fifo_time1, self.fifo_time2, self.fifo_time3, self.fifo_task_size = np.loadtxt(fname1, delimiter=",", usecols=(1,2,3,4), unpack=True)
        self.sm_time1, self.sm_time2, self.sm_time3, self.sm_task_size = np.loadtxt(fname2, delimiter=",", usecols=(1,2,3,4), unpack=True)
        self.smft_time1, self.smft_time2, self.smft_time3, self.smft_task_size = np.loadtxt(fname3, delimiter=",", usecols=(1,2,3,4), unpack=True)
        self.sum_of_all_data = np.sum(np.loadtxt(data, delimiter=",", usecols=(1,), skiprows=1, unpack=True))
        
        # self.locs = 2 * np.arange(1, len(work_time)+1)
        # self.width = 0.4

    def plot_task_exec_hist(self):
        plt.hist(self.fifo_time3 - self.fifo_time1, label="FIFO", bins=25, alpha=0.7)
        plt.hist(self.sm_time3 - self.sm_time1, label="Smart", bins=25, alpha=0.1)
        plt.hist(self.smft_time3 - self.smft_time1, label="SMFT", bins=25)
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.20),ncol=3, fancybox=True, shadow=True)
        plt.xlabel("Time")
        plt.ylabel("Amount of Task")
        plt.show
    
    def plot_sending_time(self):
        plt.hist(self.fifo_time2 - self.fifo_time1, label="FIFO", bins=25, alpha=0.7)
        plt.hist(self.sm_time2 - self.sm_time1, label="Smart", bins=50, alpha=0.1)
        plt.hist(self.smft_time2 - self.smft_time1, label="SMFT", bins=25)
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.20),ncol=3, fancybox=True, shadow=True)
        plt.xlabel("Time")
        plt.ylabel("Amount of Task")
        plt.title("Sending task hist")
        plt.show
          
    def plot_100(self):
    
        fifo_sum_task_size = np.sum(self.fifo_task_size)
        sm_sum_task_size = np.sum(self.sm_task_size)
        smft_sum_task_size = np.sum(self.smft_task_size)
        
        fifo_cumsum = np.cumsum(self.fifo_task_size) / self.sum_of_all_data
        sm_cumsum = np.cumsum(self.sm_task_size) / self.sum_of_all_data
        smft_cumsum = np.cumsum(self.smft_task_size) / self.sum_of_all_data
        
        plt.plot(self.fifo_time3, fifo_cumsum, label="FIFO", lw="2")
        plt.plot(self.sm_time3, sm_cumsum, label="Smart", lw="2")
        plt.plot(self.smft_time3, smft_cumsum, label="SMFT", lw="2")
        
        plt.plot([0, self.fifo_time3[-1]], [fifo_cumsum[-1], fifo_cumsum[-1]], "-.", color="#ff7f50")
        plt.plot([0, self.sm_time3[-1]], [sm_cumsum[-1], sm_cumsum[-1]], "-.", color="#7fffd4")
        
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.20),ncol=3, fancybox=True, shadow=True)
        plt.xlabel("Time")
        plt.ylabel("Procent of Execution of All Tasks")
        plt.show()
    
```

```python
class Plot_Tier():
    
    def __init__(self, fname1, fname2, fname3):
    
        self.fifo_work_time, self.fifo_number_task, self.fifo_fail_task, self.fifo_download_byte, self.fifo_uploadByte = np.loadtxt(fname1, delimiter=",", skiprows=1, usecols=(0,2,4,6,8), unpack=True)
        self.sm_work_time, self.sm_number_task, self.sm_failed_task, self.sm_download_byte, self.sm_uploadByte = np.loadtxt(fname2, delimiter=",", skiprows=1, usecols=(0,2,4,6,8), unpack=True)
        self.smft_work_time, self.smft_number_task, self.smft_failed_task, self.smft_download_byte, self.smft_uploadByte = np.loadtxt(fname3, delimiter=",", skiprows=1, usecols=(0,2,4,6,8), unpack=True)
        
        self.fifo_wt_sigm, self.fifo_nt_sigm, self.fifo_ft_sigm, self.fifo_db_sigm, self.fifo_ub_sigm = np.loadtxt(fname1, delimiter=",", skiprows=1, usecols=(1,3,5,7,9), unpack=True)
        self.sm_wt_sigm, self.sm_nt_sigm, self.sm_ft_sigm, self.sm_db_sigm, self.sm_ub_sigm = np.loadtxt(fname2, delimiter=",", skiprows=1, usecols=(1,3,5,7,9), unpack=True)
        self.smft_wt_sigm, self.smft_nt_sigm, self.smft_ft_sigm, self.smft_db_sigm, self.smft_ub_sigm = np.loadtxt(fname3, delimiter=",", skiprows=1, usecols=(1,3,5,7,9), unpack=True)
        
        self.locs = 2 * np.arange(1, len(self.fifo_work_time)+1)
        self.width = 0.4
    
    def plot_traffic(self):
        
        locs = 5 * np.arange(1, len(self.fifo_work_time)+1)
        width = 0.6
        
        plt.bar(locs, self.fifo_download_byte, width=width, color='black', yerr=self.fifo_db_sigm, label="FIFO_Download")
        plt.bar(locs+width, self.fifo_uploadByte, width=width, yerr=self.fifo_ub_sigm, color = '0.75', label="FIFO_Upload")
        
        plt.bar(locs+2*width, self.sm_download_byte, width=width, color='blue',  yerr=self.sm_db_sigm, label="SM_Download")
        plt.bar(locs+3*width, self.sm_uploadByte, width=width, color='#00bfff', yerr=self.sm_ub_sigm, label="SM_Upload")
        
        plt.bar(locs+4*width, self.smft_download_byte, width=width, color = 'red', yerr=self.smft_db_sigm, label="SMFT_Download")
        plt.bar(locs+5*width, self.smft_uploadByte, width=width, color = 'orange', yerr=self.smft_ub_sigm, label="SMFT_Upload")
        #plt.xticks(locs + width*1.5, locs)
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.20),ncol=3, fancybox=True, shadow=True)
        plt.title("Traffic of the Tiers")
        plt.ylabel("Amount of Traffic")
        plt.xlabel("Tiers")
        plt.show()
        
    
    def plot_how_much_tier_works(self):
    
        plt.bar(self.locs, self.fifo_work_time / 2800, width=self.width, color='#006400', yerr=self.fifo_wt_sigm/2800, label="FIFO")
        plt.bar(self.locs+self.width, self.sm_work_time / 1150, width=self.width, color = 'blue', yerr=self.sm_wt_sigm/1150, label="Smart")
        plt.bar(self.locs+2*self.width, self.smft_work_time/1273, width=self.width, color = 'red', yerr=self.smft_wt_sigm/1273, label="SMFT")
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),ncol=3, fancybox=True, shadow=True)
        plt.xlabel("Tiers")
        plt.ylabel("Operating Time Procent")
        plt.show()
    
    
    def number_executed_task(self):
        
        plt.bar(self.locs, self.fifo_number_task, width=self.width, yerr=self.fifo_nt_sigm, color='black', label="FIFO")
        plt.bar(self.locs+self.width, self.sm_number_task, width=self.width, yerr=self.sm_nt_sigm, color='blue', label="Smart")
        plt.bar(self.locs+2*self.width, self.smft_number_task, width=self.width, yerr=self.smft_nt_sigm, color='red', label="SMFT")
        plt.xlabel("Tiers")
        plt.ylabel("Number of Executed Task")
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),ncol=3, fancybox=True, shadow=True)
        plt.show()
    
    
    def number_failed_task(self):
        plt.bar(self.locs, self.fifo_fail_task, width=self.width, yerr=self.fifo_ft_sigm, color='#a52a2a', label="FIFO")
        plt.bar(self.locs+self.width, self.sm_failed_task, width=self.width, yerr=self.sm_ft_sigm, color='#ba55d3', label="Smart")
        plt.bar(self.locs+2*self.width, self.smft_failed_task, width=self.width, yerr=self.smft_ft_sigm, color='#ffff00', label="SMFT")
        plt.xlabel("Tiers")
        plt.ylabel("Number of Failed Task")
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),ncol=3, fancybox=True, shadow=True)
        plt.show()
        
```

```python
class Plot_Anomaly():
    def __init__(self, fname1, fname2, fname3):
        self.fifo_time = np.loadtxt(fname1, delimiter=",", usecols=(1,), unpack=True)
        self.sm_time = np.loadtxt(fname2, delimiter=",", usecols=(1,), unpack=True)
        self.smft_time = np.loadtxt(fname3, delimiter=",", usecols=(1,), unpack=True)
    
    def plot_anomaly(self):
        fifo_y = np.arange(1, len(self.fifo_time)+1)
        sm_y = np.arange(1, len(self.sm_time)+1)
        smft_y = []
        for i in range(len(self.smft_time)):
            if((i % 2) == 0):
                smft_y.append(0)
            else:
                smft_y.append(1)
        plt.plot(self.fifo_time, fifo_y, color="#8b4513", lw=2.5, label="FIFO")
        plt.plot(self.sm_time, sm_y, color="#9370db", lw=2.5, label="Smart")
        plt.plot(self.smft_time, smft_y, color="#ff4500", label="SMFT")
        plt.xlabel("Time")
        plt.ylabel("Number of Anomaly")
        plt.title("Anomaly")
        plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15),ncol=3, fancybox=True, shadow=True)
        plt.show()
        
        
```
